### Q-Learning Implementation

```python
Q = np.zeros((48, 4))
```

```python
def epsilon_greedy(state):
    if random.random() < epsilon:  # exploration
        # choose random action
        return env.action_space.sample()
    else:  # exploitation
        # choose optimum action
        # from stored all Q-values
        return np.argmax(Q[state])
```

## Q-Learning Update Rule

$$
Q(s_t, a_t) \leftarrow Q(s_t, a_t)
+ \alpha \left[
r_{t+1}
+ \gamma \max_{a} Q(s_{t+1}, a)
- Q(s_t, a_t)
\right]
$$

Episode ends:
1. Terminate: agent reached at end
2. Truncate: cut short=> time limit, path length


```python
for episode in range(episodes):
    render = episode % 50 == 0

    if render:
        env = gym.make("CliffWalking-v1", render_mode="human")
        print(f"rendering env for episode = {episode+1}/{episodes}")
    else:
        env = gym.make("CliffWalking-v1")

    done = False
    state, _ = env.reset()

    episode_length = 0
    total_reward = 0

    while not done:
        action = epsilon_greedy(state)

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # Q-Learning Update
        Q[state, action] += alpha * (
            reward + gamma * np.max(Q[next_state]) - Q[state, action]
        )

        state = next_state

        total_reward += reward
        episode_length += 1

    print(
        f"Episode = {episode+1}/{episodes}: Total Reward = {total_reward} & Episode Length = {episode_length}"
    )
    env.close()
```

#### What did our agent learn?
